```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

#    
Fs = 44100 #   
T = 1/Fs #   
t = 0.1 #  
N = Fs*t #     

#    
freq = 100 #  ,    
omega = 2*np.pi*freq #    

t_vec = np.arange(N)*T #     
y = np.sin(omega*t_vec)

plt.plot(t_vec,y)
plt.show()


# fourier transform and frequency domain
#
Y_k = np.fft.fft(y)[0:int(N/2)]/N #    numpy
Y_k[1:] = 2*Y_k[1:] #      
Pxx = np.abs(Y_k) # '   

f = Fs*np.arange((N/2))/N; #  

# plotting
fig,ax = plt.subplots()
plt.plot(f,Pxx,linewidth=5)
ax.set_xscale('log')
ax.set_yscale('log')
plt.ylabel('Amplitude')
plt.xlabel('Frequency [Hz]')
plt.show()
```

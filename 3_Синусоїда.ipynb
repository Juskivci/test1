```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# інформація про вибірку 
Fs = 44100 # частота дискретизації 
T = 1/Fs # період вибірки 
t = 0.1 # секунди вибірки
N = Fs*t # загальна кількість точок у сигналі

# інформація про сигнал 
freq = 100 # в герцах, бажана власна частота 
omega = 2*np.pi*freq # кутова частота для синусоїди

t_vec = np.arange(N)*T # вектор часу для побудови графіка
y = np.sin(omega*t_vec)

plt.plot(t_vec,y)
plt.show()


# fourier transform and frequency domain
#
Y_k = np.fft.fft(y)[0:int(N/2)]/N # функція ШПФ з numpy
Y_k[1:] = 2*Y_k[1:] # потрібно візьміть лише односторонній спектр 
Pxx = np.abs(Y_k) # обов'язково позбудьтеся уявної частини

f = Fs*np.arange((N/2))/N; # частотний вектор

# plotting
fig,ax = plt.subplots()
plt.plot(f,Pxx,linewidth=5)
ax.set_xscale('log')
ax.set_yscale('log')
plt.ylabel('Amplitude')
plt.xlabel('Frequency [Hz]')
plt.show()
```
